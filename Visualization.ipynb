```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("C:\My\Data.csv")
data.head()
```

```python
df=data[['Shots','OnTarget']][:50]
df.plot(title='Shots', style = ['g','r'])
plt.xlabel('Player')
plt.ylabel('Summ of shots');
```

```python
df1=data[['Matches_Played','Shots','Goals','OnTarget']][:50]
ax = df1.plot(style=['r-', 'g--', 'b:', 'm-.'], title='Statistics');
```

```python
New_data=data[data['Year'] == 2016]
New_data.set_index('Player Names', inplace = True)
New_data.head()
```

```python
df2 = New_data[['Shots','OnTarget']][:10]
df2.plot(title='Shots', kind = 'bar', color = ['b', 'r']);
```

```python
df3 = New_data[['On Target Per Avg Match','Shots Per Avg Match']][:10]
df3.plot(title='Per Avg Match', kind = 'barh', stacked = True, color = ['y','m']);
```

```python
df3.boxplot(return_type = 'axes')
```

```python
df4 = New_data[['Matches_Played', 'Goals']][:5]
ax = df4.plot(kind='area', stacked = False)
xticklabels = data['Player Names'][:11]
ax.set_xticklabels(["\n"*(i%2) + l for i,l in enumerate(xticklabels)])
plt.setp(ax.get_xticklabels(), rotation=45, ha="right");
```

```python
df5 = New_data[['Matches_Played', 'Goals']]
df5.plot(kind = 'scatter', x = 'Matches_Played', y = 'Goals', c = 'm');
```

```python
df6=New_data[['Matches_Played','OnTarget','Shots']][:10]
plt.imshow(df6, cmap='hot', interpolation='none')
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=10)
plt.xticks(range(len(df6.columns)), df6.columns,size = 10,rotation=90) 
plt.yticks(range(len(df6)), df6.index, size = 10);
```

```python
df7=New_data.nlargest(10,['Goals'])[['Goals']]
df7.plot(kind='bar', title='Top scorers', color='r');
plt.ylabel('Goals')
plt.grid(visible=True, axis='y', linestyle='--', color='c');
```
